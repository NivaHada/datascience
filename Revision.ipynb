````python
class MyObject():
    def __init__(self):
        self.data = 7
        self.count = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        self.count += 1
        if self.count <= 1:
            return self.data
        else:
            raise StopIteration()
````

````python
raise StopIteration()
````

````python
for item in MyObject():
    print(item)
````

````python
import time
````

````python
def log(func, *args, **kwargs):
    start = time.time()
    result = func.__call__(*args, **kwargs)
    print("Function {} took {} seconds".format(func.__name__, time.time() - start))
    return result
````

````python
def fib(a, b):
    result = []
    while b < 10000:
        a, b = b, a + b
        result.append(b)
    return result
````

````python
fib(0, 1)
````

````python
log(fib, 0, 1)
````

````python
def log(func):
    def _wrapper(*args, **kwargs):
        start = time.time()
        result = func.__call__(*args, **kwargs)
        print("Function {} took {} seconds".format(func.__name__, time.time() - start))
        return result
    return _wrapper
````

````python
logged_fib = log(fib)
````

````python
logged_fib
````

````python
logged_fib(0, 1)
````

````python
fib(0, 1)
````

````python
fib = log(fib)
````

````python
fib
````

````python
fib(0, 1)
````

````python
@log
def fib22(a, b):
    result = []
    while b < 10000:
        a, b = b, a + b
        result.append(b)
    return result
````

````python
fib22(0, 1)
````

````python
fib
````

GIL - Global Interpreter Lock

````python
import threading
````

[https://docs.python.org/3/library/threading.html](https://docs.python.org/3/library/threading.html)

````python
import asyncio
````

````python
@asyncio.coroutine
def todo(task, given_time=3):
    print("Task {} started with {} seconds".format(task, given_time))
    yield from asyncio.sleep(given_time)
    print("Task {} completed after {} seconds".format(task, given_time))
````

````python
loop = asyncio.get_event_loop()
````

````python
task = [
    todo("A", 4),
    todo("B", 2),
    todo("C", 5),
    todo("D", 2)
]
````

````python
loop.run_until_complete(asyncio.wait(task))
````

````python
loop.close()
````
